# 1.6 awk

This notebook will demonstrate the basics of using `awk` to extract specific fields from within data files.

```python
%%bash
awk -F "\t" '{print NF}' data/art/artists.txt
```

```python
%%bash
awk -F "\t" '{print $3}' data/art/artists.txt
```

```python
%%bash
awk -F "\t" '{print $3}' data/art/artists.txt | sort | uniq -c | sort -k1,1nr | head -1
```
