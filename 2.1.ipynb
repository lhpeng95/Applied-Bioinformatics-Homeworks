```r
#for Question 3
primelist <- integer(500)
primelist[1] <- 2
cont <- 1          #cont for prime number
num <- 3
while(cont <500){           #cont current prime#
    flag <- TRUE           #indicator of whether prime
    i <- 1
    while(primelist[i]<=sqrt(num)&&i<cont+1){
        if (num%%primelist[i]==0){
            flag <- FALSE
            break
        }
        i <- i+1
    }
    if(flag==TRUE){
        cont <- cont+1
        primelist[cont] <-num
    }
    num <- num +2
}
primelist
print(paste("The 500th prime is ",primelist[500]))
```

```r
#for question 5
#5.1 Subset both dataframes to include only "baseMean", "log2FoldChange" and "padj"
#5.2 Merge the two dataframes (Change column names so that there are no repetative column names. The base mean in two spreadsheets are the same so delete the extra one.) into a dataframe df.nav.d14.mem.
#5.3 Sort the new dataframe on " baseMean".
#5.4 Define a function to evaluate padj output:
#- padj > 0.05: nonsig
#- padj <= 0.05: sig
#5.5 Apply the function to both nav.mem and nav.d14 and create one column to add the evaluation result.

setwd("/home/lhpeng/Applied-Bioinformatics/Unit1-module2-R")
df.nav.d14 <- read.csv("NAV-D14_DEseq2.csv", header=TRUE)
df.nav.mem <- read.csv("NAV-MEM_DEseq2.csv", header=TRUE)
sub.nav.d14 <- df.nav.d14[c("baseMean", "log2FoldChange", "padj")]
sub.nav.mem <- df.nav.mem[c("baseMean", "log2FoldChange", "padj")]
sub.nav <- rbind(sub.nav.d14, sub.nav.mem)
sub.nav.sort <- sub.nav[order(sub.nav$baseMean),]
padj.ev <- function(p){
    if(is.na(p)){
        return (NA)
    }
    else if(p>0.05){
        return ("nonsig")
    }
    else{
        return("sig")
    }
}
padjev.nav <- lapply(sub.nav.sort$padj, padj.ev)
sign <- unlist(padjev.nav)
cbind(sub.nav.sort, sign)
```
